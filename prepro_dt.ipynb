```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
dataset= pd.read_csv('mba.csv')
dataset.shape
```

```python
dataset.head(10)
```

```python
dataset.drop(columns= ["application_id"])
```

```python
dataset.isnull().sum()
```

```python
dataset['race'].fillna('International', inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.drop(columns= ["international"])
```

```python
dataset.isnull().sum()
```

```python
dataset.replace('Admit', '1', inplace=True)
dataset.head(15)
```

```python
dataset.replace('Waitlist', '1', inplace=True)
dataset.head(15)
```

```python
dataset['admission'].fillna('0', inplace=True)
dataset.head(9)
```

```python
label_encoders = {}
categorical_cols = ["major", "work_industry", "gender", "race"]
for col in categorical_cols:
    le = LabelEncoder()
    dataset[col] = le.fit_transform(dataset[col])
    label_encoders[col] = le
```

```python
dataset['admission'] = pd.to_numeric(dataset['admission'])
dataset.head(5)
```

```python
y = dataset["admission"]
x = dataset.drop(columns= ["admission"])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
dataset.head(5)
```

```python
y.head(5)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
y_pred_dt = dt.predict(X_test)
report_dt = classification_report(y_test, y_pred_dt)
print(report_dt)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize=(16, 10))
plot_tree(dt, feature_names=x.columns, class_names=["Rejected", "Admitted"], filled=True)
plt.title("Decision Tree Visualization")
plt.show()
```

```python
from sklearn.metrics import accuracy_score
print('accuracy ', accuracy_score(y_pred_dt,y_test))
```
