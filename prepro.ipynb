```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
dataset= pd.read_csv('mba.csv')
dataset.shape
```

```python
dataset.head(10)
```

```python
dataset.drop(columns= ["application_id"])
```

```python
dataset.isnull().sum()
```

```python
dataset['race'].fillna('International', inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.drop(columns= ["international"])
```

```python
dataset.isnull().sum()
```

```python
dataset.replace('Admit', '1', inplace=True)
dataset.head(15)
```

```python
dataset.replace('Waitlist', '1', inplace=True)
dataset.head(15)
```

```python
dataset['admission'].fillna('0', inplace=True)
dataset.head(9)
```

```python
label_encoders = {}
categorical_cols = ["major", "work_industry", "gender", "race"]
for col in categorical_cols:
    le = LabelEncoder()
    dataset[col] = le.fit_transform(dataset[col])
    label_encoders[col] = le
```

```python
dataset['admission'] = pd.to_numeric(dataset['admission'])
dataset.head(5)
```

```python
y = dataset["admission"]
x = dataset.drop(columns= ["admission"])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```
